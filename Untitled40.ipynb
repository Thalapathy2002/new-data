```python
import os
import pandas as pd
import torch
import easyocr
from PIL import Image
from transformers import pipeline
import matplotlib.pyplot as plt

# Step 1: Load images
image_folder = 'images'
image_paths = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

df = pd.DataFrame({'image_path': image_paths})
df['image_name'] = df['image_path'].apply(lambda x: os.path.basename(x))
```

```python
reader = easyocr.Reader(['en'])

def extract_text(image_path):
    try:
        results = reader.readtext(image_path, detail=0)
        return " ".join(results)
    except:
        return ""

df['ocr_text'] = df['image_path'].apply(extract_text)
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")

def predict_sentiment(text):
    if not text.strip():
        return "neutral"
    return sentiment_pipeline(text)[0]['label'].lower()

df['predicted_sentiment'] = df['ocr_text'].apply(predict_sentiment)
```

```python
for i in range(min(5, len(df))):
    row = df.iloc[i]
    img = Image.open(row['image_path'])
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Predicted: {row['predicted_sentiment']}\nOCR: {row['ocr_text'][:80]}")
    plt.show()
```

```python
df.to_csv("results_from_images_only.csv", index=False)
```

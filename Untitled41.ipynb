```python
# Step 1: Imports
import os
import pandas as pd
import torch
import easyocr
import numpy as np
from PIL import Image
from transformers import pipeline
from torchvision import transforms, models
import matplotlib.pyplot as plt

# Step 2: Load CSV Dataset
try:
    df = pd.read_csv('labels.csv')  # Should contain 'image_name', 'sentiment'
    print("Loaded labels.csv with", len(df), "entries.")
except Exception as e:
    print("labels.csv not found or error loading:", e)
    df = pd.DataFrame()

# Step 3: Load Images Folder
image_folder = 'images'
image_paths = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Step 4: EasyOCR for Text Extraction
reader = easyocr.Reader(['en'])

def extract_text(image_path):
    try:
        result = reader.readtext(image_path, detail=0)
        return " ".join(result)
    except:
        return ""

# Add OCR text to DataFrame or create new
if not df.empty:
    df['image_path'] = df['image_name'].apply(lambda x: os.path.join(image_folder, x))
else:
    df = pd.DataFrame({'image_path': image_paths})
    df['image_name'] = df['image_path'].apply(lambda x: os.path.basename(x))

df['ocr_text'] = df['image_path'].apply(extract_text)

# Step 5: Hugging Face BERT Sentiment Analysis on OCR Text
sentiment_pipeline = pipeline("sentiment-analysis")

df['predicted_sentiment'] = df['ocr_text'].apply(lambda x: sentiment_pipeline(x)[0]['label'] if x else "neutral")

# Step 6: Show Results
for i in range(min(5, len(df))):  # Show 5 samples
    row = df.iloc[i]
    img = Image.open(row['image_path'])
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Predicted: {row['predicted_sentiment']}\nOCR: {row['ocr_text'][:80]}")
    plt.show()

# Step 7: Compare with True Labels (if available)
if 'sentiment' in df.columns:
    from sklearn.metrics import classification_report
    true = df['sentiment'].astype(str).str.lower()
    pred = df['predicted_sentiment'].astype(str).str.lower()
    print("Classification Report:\n", classification_report(true, pred))
```

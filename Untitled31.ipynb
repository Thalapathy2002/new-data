```python
2) #Generate and plot random data from two Gaussian distributions: one with a mean of 10 and
a standard deviation of 2, and the other with a mean of 15 and a standard deviation of 5.
What differences do you observe in the histograms?
    
    
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_and_plot_two_gaussian_samples(mean1=10, std_dev1=2, mean2=15, std_dev2=5, num_samples=5000):
    # Generate random samples from the first Gaussian distribution
    samples1 = np.random.normal(loc=mean1, scale=std_dev1, size=num_samples)
    
    # Generate random samples from the second Gaussian distribution
    samples2 = np.random.normal(loc=mean2, scale=std_dev2, size=num_samples)
    
    # Plot the histogram of the first set of samples
    plt.hist(samples1, bins=50, density=True, alpha=0.6, color='b', label=f'Mean={mean1}, Std Dev={std_dev1}')
    
    # Plot the histogram of the second set of samples
    plt.hist(samples2, bins=50, density=True, alpha=0.6, color='r', label=f'Mean={mean2}, Std Dev={std_dev2}')
    
    plt.title('Histograms of Two Gaussian Distributions')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Call the function
generate_and_plot_two_gaussian_samples()
```

```python
1) #Write a Python function to generate 5000 random samples from
a Gaussian distribution with a mean of 50 and a standard deviation of 5.
Plot the histogram of the data.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_and_plot_gaussian_samples(mean=50, std_dev=5, num_samples=5000):
    # Generate random samples from a Gaussian distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
    
    # Plot the histogram of the samples
    plt.hist(samples, bins=50, density=True, alpha=0.6, color='g')
    
    # Plot the probability density function (PDF)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = np.exp(-((x - mean) ** 2) / (2 * std_dev ** 2)) / (std_dev * np.sqrt(2 * np.pi))
    plt.plot(x, p, 'k', linewidth=2)
    
    plt.title('Histogram of Gaussian Samples')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

# Call the function
generate_and_plot_gaussian_samples()
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score

# Reading data from URLs
confirmed_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
latest_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01-15-2023.csv')

# Extracting confirmed cases up to January 15, 2023
confirmed_cases_latest = latest_data['Confirmed']

# Extracting true confirmed cases from the previous dataset
true_confirmed_cases = confirmed_data.iloc[:, 4:].sum(axis=0)

# Converting index to datetime format for consistency
true_confirmed_cases.index = pd.to_datetime(true_confirmed_cases.index)

# Calculating accuracy score
accuracy = accuracy_score(true_confirmed_cases, confirmed_cases_latest)

print("Accuracy Score:", accuracy)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score

# Reading data from URLs
confirmed_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
latest_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01-15-2023.csv')

# Extracting confirmed cases up to January 15, 2023
confirmed_cases_latest = latest_data['Confirmed']

# Extracting true confirmed cases from the previous dataset
true_confirmed_cases = confirmed_data.iloc[:, 4:].sum(axis=1)  # Corrected axis to sum across columns

# Convert true confirmed cases to a 1D array
true_confirmed_cases_flat = true_confirmed_cases.values.flatten()

# Calculating accuracy score
accuracy = accuracy_score(true_confirmed_cases_flat, confirmed_cases_latest)

print("Accuracy Score:", accuracy)
```

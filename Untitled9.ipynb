```python
#Write a Python program to execute the given matrix x = [2, 3, 4, 7], [0, 1, 5, 3], [4, 4, 2, 0], [1, 8, 6, 3]
import numpy as np
a = np.array([[2, 3, 4, 7], [0, 1, 5, 3], [4, 4, 2, 0], [1, 8, 6, 3]])
print("The eigen values are:\n", a)
w,v = np.linalg.eig(a)
print("E-value", w)
print("E-vector", v)
```

```python
# Write a Python program to fill null values with the previous values 
#Previous is of two types:
# Forward null values - 'ffill'
# Backward null values - 'bfill'
import pandas as pd
data = {'A': [None, 10, 17, None],
        'B': [30, 4, None, None],
        'C': [10, None, 4, None]};
df = pd.DataFrame(data)
df_filled = df.fillna(method='ffill');

print(df)
print(df_filled)
```

```python
# Write a Python program using interpolate() function to fill the missing values using Linear method
import pandas as pd
import numpy as np
data = {'A': [1, np.nan, 3, np.nan, 5],
        'B': [np.nan, 2, np.nan, 4, np.nan]};

df = pd.DataFrame(data)

print(df)
```

```python
# Interpolate() function
# Q. Write a python program using interpolate function to fill the missing valueses using linear method 
import pandas as pd
data = {
        'X:' : [1, None, 3 ,None,5],
        'Y:' : [None, 2, 4, None,7]
        
}
df = pd.DataFrame(data)
print("Previous data : \n", df)
df_interpolate = df.interpolate(method='linear')
print("Filled Data - linear: \n", df_interpolate)
```

```python
#Write a Python program using dropped function to print atleast one null value
import pandas as pd
data = {'a': [1, None, 3, None, 5],
        'b': [None, 2, 4, None, 7]}
df = pd.DataFrame(data)
df_dropped = df.dropna()
print(df)
print(df_dropped)
```

```python
#Write a Python program using dropped function to print atleast one null value
import pandas as pd
data = {'a': [1, None, 3, None, 5],
        'b': [None, 2, 4, None, 7]}
df = pd.DataFrame(data)
df_dropped = df.dropna(axis=1)
print(df)
print(df_dropped)
```

```python
#Write a Python program using dropped function to print atleast one null value
import pandas as pd
data = {'a': [1, None, 3, None, 5],
        'b': [None, 2, 4, None, 7]}
df = pd.DataFrame(data)
df_dropped = df.dropna(how='all')
print(df)
print(df_dropped)
```

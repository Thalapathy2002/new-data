```python
#Write inpython program to find if A and B are two fuzzy sets with membership functions A(x) = {0.2, 0.5, 0.6, 0.1, 0.9} B(x) = {0.1, 0.5, 0.2, 0.7, 0.8} Then the value of A intersection B.''


def fuzzy_intersection(A, B):
    intersection = [min(A[i], B[i]) for i in range(len(A))]
    return intersection

A = [0.2, 0.5, 0.6, 0.1, 0.9]
B = [0.1, 0.5, 0.2, 0.7, 0.8]
intersection = fuzzy_intersection(A, B)
print("Intersection of A and B:", intersection)
```

```python
def fuzzy_union(mu_a, mu_b):
    assert len(mu_a)==len(mu_b),"Fuzzy sets must have same length"
    union = [max(a,b)for a,b in zip(mu_a,mu_b)]
    return union

mu_a = [0.2, 0.5, 0.6, 0.1, 0.9]
mu_b = [0.1, 0.5, 0.2, 0.7, 0.8]
union_result = fuzzy_union(mu_a, mu_b)
print("Union of mu_a and mu_b:", union_result)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# First dataset
data1 = {'counter': ["A", "B", "C", "D", "E"],
         'per_capita_index': [45000, 42000, 52000, 49000, 47000]}
df1 = pd.DataFrame(data1)

# Second dataset
data2 = {'x_axis': ['value_1', 'value_2', 'value_3'],
         'y_axis': [10, 20, 15]}  # Example numerical values
df2 = pd.DataFrame(data2)

# Plotting
fig, ax1 = plt.subplots()

# Plotting bar chart for df1
ax1.bar(df1['counter'], df1['per_capita_index'], color='b')
ax1.set_xlabel('Counter')
ax1.set_ylabel('Per Capita Index', color='b')

# Creating a secondary y-axis for df2
ax2 = ax1.twinx()
ax2.plot(df2['x_axis'], df2['y_axis'], color='r', marker='o')
ax2.set_ylabel('Y Axis Name', color='r')

plt.title('Title Name')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Given dataset
students = ['Raghav', 'Sagar', 'Jyothir', 'Parthiban', 'Joseph', 'Daniel']
score_card = [65, 47, 83, 37, 77, 89]

# Bar chart
plt.figure(figsize=(8, 6))
plt.bar(students, score_card, color='skyblue')
plt.title('Student Scores')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.xticks(rotation=45)
plt.show()

# Line graph
plt.figure(figsize=(8, 6))
plt.plot(students, score_card, marker='o', color='green')
plt.title('Student Scores')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(score_card, labels=students, autopct='%1.1f%%', startangle=140)
plt.title('Student Scores Distribution')
plt.axis('equal')
plt.show()
```

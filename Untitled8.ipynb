```python
#Write a python program to execute the given matrix a=[0, 2, 2, 3]
import numpy as np
from numpy.linalg import eig
a = np.array([[0, 2], [2, 3]])
w,v=eig(a)
print('E-value', w)
print('E-vector', v)
```

```python
#Write a python python to compute eigenvlaues and eigenvectors a = [3, 7, 3, 4, 1, 3, 1, 4, 6]
import numpy as np
from numpy.linalg import eig
a = np.array([[3, 7, 3], [4, 1, 3], [1, 4, 6]])
c,d = eig(a)
print("E-value", c)
print("E-vector", d)
```

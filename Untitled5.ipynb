```python
import pandas as pd
#Data set loading
wq=pd.read_csv("winequalityRED.csv")
wq.info()
```

```python
#checking for the null values 
wq.isnull().sum()
```

```python
#replacing missing values with mean values 
mean_density = wq["density"].mean()
mean_chlorides = wq["chlorides"].mean()
print("Mean density is", mean_density)
print("Mean chlorides", mean_chlorides)
```

```python
wq["density"].fillna(value=mean_density, inplace=True)
wq["chlorides"].fillna(value=mean_chlorides, inplace=True)
```

```python
#Exporting the dataset
wq.to_csv("clean_file.csv")
```

```python
import random

def two_point_crossover(parent1, parent2):
    # Randomly select two crossover points
    crossover_points = sorted(random.sample(range(1, len(parent1)), 2))
    crossover_point1, crossover_point2 = crossover_points
    
    # Create children by combining segments from parents
    child1 = parent1[:crossover_point1] + parent2[crossover_point1:crossover_point2] + parent1[crossover_point2:]
    child2 = parent2[:crossover_point1] + parent1[crossover_point1:crossover_point2] + parent2[crossover_point2:]
    
    return child1, child2

# Example individuals
x1 = [6, 5, 4, 1, 3, 5, 3, 2]
x2 = [8, 7, 1, 2, 6, 6, 0, 1]
x3 = [2, 3, 9, 2, 1, 2, 8, 5]
x4 = [4, 1, 8, 5, 2, 0, 9, 4]

# Perform crossover between x1 and x2
child1_x1_x2, child2_x1_x2 = two_point_crossover(x1, x2)
print("Two-point crossover between x1 and x2:")
print("Child 1:", child1_x1_x2)
print("Child 2:", child2_x1_x2)
print()

# Perform crossover between x3 and x4
child1_x3_x4, child2_x3_x4 = two_point_crossover(x3, x4)
print("Two-point crossover between x3 and x4:")
print("Child 1:", child1_x3_x4)
print("Child 2:", child2_x3_x4)
```
